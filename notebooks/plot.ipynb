```python
from matplotlib import pyplot as plt
```

```python
temperatures = [1, 5e-1, 1e-1, 5e-2, 1e-2, 5e-3, 1e-3]
```

```python
errors_laplace_fcnn = [0.0345, 0.03537, 0.03603, 0.03772, 0.03649, 0.03626, 0.03581]
errors_laplace_cnn = [0.01766, 0.0164, 0.01644, 0.01892, 0.01683, 0.01687, 0.01801]
```

```python
errors_normal_fcnn = [0.03461, 0.03194, 0.03362, 0.03362, 0.03085, 0.02996, 0.03076]
errors_normal_cnn = [0.0179, 0.01839, 0.0179, 0.01731, 0.0178, 0.01701, 0.01711]
```

```python
errors_cnn_sgd = [1 - 0.9822] * 7
errors_fcnn_sgd = [1 - 0.9521] * 7
```

```python
plt.figure(figsize = (20, 20))
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 3))
ax1.set_xscale('log')
ax1.plot(temperatures, errors_laplace_fcnn, label = 'laplace')
ax1.plot(temperatures, errors_normal_fcnn, label = 'normal')
ax1.plot(temperatures, errors_fcnn_sgd, label = 'baseline', linestyle = '--', color = 'grey')
ax1.legend()
ax1.set_title(label = 'FCNN', fontsize = 20)
ax2.set_title('CNN', fontsize = 20)
ax1.set_ylabel('MNIST', fontsize = 20)
ax2.set_xscale('log')
ax2.plot(temperatures, errors_laplace_cnn, label = 'laplace')
ax2.plot(temperatures, errors_normal_cnn, label = 'normal')
ax2.plot(temperatures, errors_cnn_sgd, label = 'baseline', linestyle = '--', color = 'grey')
ax2.legend()
plt.savefig('errors.png')
plt.show()
```
